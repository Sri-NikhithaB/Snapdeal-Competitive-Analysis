```python
df.set_index('Unique ID for matched products', inplace=True)
```

```python
# Define the list of columns to check for -1 or 0
columns_to_check = ['Jan 8 2024', 'Jan 7 2024', 'Jan 6 2024', 'Jan 5 2024', 'Jan 4 2024', 'Jan 3 2024', 'Jan 2 2024', 'Jan 1 2024']

# Create a boolean mask to identify rows where any of the specified columns contain -1 or 0
mask = df[columns_to_check].isin([-1, 0]).any(axis=1)

# Drop the rows identified by the mask
df.drop(df[mask].index, inplace=True)
df.head()
```

```python
# Replace 'Snap Deal_app' with 'Snap Deal' in the 'Source' column
df['Source'] = df['Source'].replace('Snap Deal_app', 'Snap Deal')
```

```python
# Replace values in the 'Stock' column
df['Stock'] = df['Stock'].replace(['In Stock.', 'instock', 'Available'], 'In Stock')
df['Stock'] = df['Stock'].replace(['out of stock', 'Out of Stock!', 'Out of Stock.'], 'Out of Stock')
df['Stock'] = df['Stock'].replace(['In Stock - only 1 left', 'In Stock - only 2 left', 'In Stock - only 3 left', 'In Stock - only 4 left', 'In Stock - only 5 left', 'In Stock - only 6 left', 'In Stock - only 7 left', 'In Stock - only 8 left'], 'In Stock - only few left')
df['Stock'] = df['Stock'].replace(['Discontinued.', 'No Sellers Available', '^\s*$'], 'NA')

# Fill remaining NaN values with 'NA'
df['Stock'].fillna('NA', inplace=True)
```

```python
# Replace values in the 'Category' column
df['Category'] = df['Category'].replace(['^\s*$'], 'NA')
```

```python
unique_values = df['Category'].unique()
print(unique_values)
```

```python
# Replace values in the 'Category' column
df['Category'] = df['Category'].replace(['TVS,AUDIO & VIDEO','TV Shop', 'TVs, Audio & Video', 'TVs, Audio \\u0026 Video',], 'TVs Audio & Video')
df['Category'] = df['Category'].replace(['Laptop Batteries', 'Computer Peripherals','Computers \\u0026 Peripherals', 'Pen Drives', 'computers-mouse', 'computers-batteries', 'Computers & Peripherals','computers-pen-drives', 'computers-components'], 'Computers & Components')
df['Category'] = df['Category'].replace(['Gaming','gaming-titles'], 'Gaming')
df['Category'] = df['Category'].replace(['Mobiles \\u0026 Tablets', 'Mobiles & Accessories', 'mobiles-power-banks', 'Mobiles & Tablets'], 'Mobiles & Accessories')
df['Category'] = df['Category'].replace(['Camera Accessories','Cameras \\u0026 Accessories', 'cameras-lenses', 'Cameras & Accessories', 'cameras-camera-accessories'], 'Cameras & Accessories')
df['Category'] = df['Category'].replace(['computers-speakers', 'Headphones & Mics','electronic-headphones-earphones', 'computers-headphones-mics'], 'Electronics')
df['Category'] = df['Category'].replace(['Health, Wellness & Medicine', 'health-heart-rate-bp-monitor'], 'Health & Wellness')
df['Category'] = df['Category'].replace(['cameras-binoculars-telescopes'], 'Binoculars & Telescopes')
```

```python
unique_values = df['Category'].unique()
print(unique_values)
```

```python
# Replace values in the 'Brand' column
df['Brand'] = df['Brand'].replace(['^\s*$'], 'NA')
```

```python
# Calculate the maximum selling price across all date columns
df['Original Price'] = df[['Jan 8 2024', 'Jan 7 2024', 'Jan 6 2024', 'Jan 5 2024', 'Jan 4 2024', 'Jan 3 2024', 'Jan 2 2024', 'Jan 1 2024']].max(axis=1)

# Display the DataFrame with the new "Original Price" column
df.head()
```

```python
# Calculate discount for each day using selling price and original price
for day in ['Jan 8 2024', 'Jan 7 2024', 'Jan 6 2024', 'Jan 5 2024', 'Jan 4 2024', 'Jan 3 2024', 'Jan 2 2024', 'Jan 1 2024']:
    df[f'Discount_{day}'] = (df['Original Price']-df[day] ) * 100 / df['Original Price']

# Find the maximum discount among the discounts for the 8 days
df['Max Discount'] = df.filter(like='Discount_').max(axis=1)
df
```

```python
discounted_df = df[df['Max Discount'] !=0]
discounted_df
```

```python
# Group by 'Category' and find the index of the maximum discount within each category
max_discount_indices = discounted_df.groupby('Category')['Max Discount'].idxmax()

# Use the indices to retrieve the corresponding 'Source' and 'Discount' values
max_discount_info = discounted_df.loc[max_discount_indices, ['Category', 'Source', 'Max Discount']]

# Resetting index for better representation if needed
max_discount_info.reset_index(drop=True, inplace=True)

# Add a column to show the maximum discount offered by Snapdeal for each category
def get_snapdeal_max_discount(category):
    snapdeal_discounts = discounted_df[(discounted_df['Category'] == category) & (discounted_df['Source'] == 'SnapDeal')]
    if not snapdeal_discounts.empty:
        return snapdeal_discounts['Max Discount'].max()
    else:
        return 0

max_discount_info['Snapdeal_Discount'] = max_discount_info['Category'].apply(get_snapdeal_max_discount)

# Print or display the DataFrame
print(max_discount_info)
```

```python
max_discount_info.drop(max_discount_info[max_discount_info['Source'] == 'SnapDeal'].index, inplace=True)
max_discount_info.reset_index(drop=True, inplace=True)
print(max_discount_info)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded the dataframe
# df = <your_dataframe>

# Calculate the difference between Snapdeal's discount and other sources' maximum discounts
max_discount_info['Difference'] = max_discount_info['Snapdeal_Discount'] - max_discount_info['Max Discount']

# Group by Source and calculate average difference
avg_difference_by_source = max_discount_info.groupby('Source')['Difference'].mean().reset_index()

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(avg_difference_by_source['Source'], avg_difference_by_source['Difference'], color='skyblue')
plt.xlabel('Source')
plt.ylabel('Average Difference in Discounts (%)')
plt.title('Comparison of Snapdeal Discount with Other Sources')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)  # Add zero line for reference
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded the dataframe
# max_discount_info = <your_dataframe>

# Calculate the difference between Snapdeal's discount and other sources' maximum discounts
max_discount_info['Difference'] = max_discount_info['Snapdeal_Discount'] - max_discount_info['Max Discount']

# Group by Category and Source, then calculate average difference
avg_difference_by_category = max_discount_info.groupby(['Category', 'Source'])['Difference'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 8))

# Iterate over each category
for category, data in avg_difference_by_category.groupby('Category'):
    plt.bar(data['Source'], data['Difference'], label=category)

plt.xlabel('Source')
plt.ylabel('Average Difference in Discounts (%)')
plt.title('Comparison of Snapdeal Discount with Other Sources by Category')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.5)  # Add zero line for reference
plt.xticks(rotation=45)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
